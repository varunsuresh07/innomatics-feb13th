```python
#problem 1
Parking_Capacity=50
Vehicle_Logs=["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
def val(x,y):
  count=0
  for i in y:
    if i=="IN":
      count+=1
    else:
      count=count-1
  print("Crrently Parked Vehicles:",count)
  if count<x:
    print("Parking Status:Available")
  else:
    print("Parking Status: Full")


val(Parking_Capacity,Vehicle_Logs)
```

```python
#problem 2
def cal(dis,lel,weat):
  speed=30
  time=dis/speed
  delay=0
  if lel=="low":
    delay+=5
  elif lel=="Medium":
    delay+=10
  elif lel=="High":
    delay+=20
  if weat=="Rainy":
    delay+=10
  elif weat=="Snowing":
    delay+=20
  tim=time+delay
  print("Estimated Delivery Time:",tim)
Distance= 8
Traffic="High"
Weather="Rainy"
cal(Distance,Traffic,Weather)

```

```python
#problem 5
# Problem 5: Smart Classroom Resource Usage Monitor

def monitor_resources():
    resource_usage = {
        "Projector": 6,
        "AC": 9,
        "Lights": 4
    }

    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

monitor_resources()
```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_registration(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Loop through each registration
    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

manage_registration(capacity, registrations)
```
